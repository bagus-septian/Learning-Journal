## **Task 4**

Gunakan for loop dan conditional if untuk membuat sebuah list yang berisikan penambahan jumlah kasus harian Covid-19 dengan ketentuan hanya mengambil data Indonesia saja dan jumlah kasus di bawah 10000. Berapa jumlah hari dengan kasus di bawah 10000? (Gunakan for loop dan conditional if dalam hal ini)

```python
df_harian.reset_index(drop=True, inplace=True)
```

```python
df_harian.shape
```

```python
jumlah = []

for i in range(0, len(df_harian)):
    if df_harian.loc[i, "Indonesia"] < 10000:
        jumlah.append(df_harian.loc[i, "Indonesia"])
    elif df_harian.loc[i, "Indonesia"] > 10000:
        continue

jumlah
```

```python
len(jumlah)
```

Jumlah penambahan kasus harian pada dataframe yang berjumlah kurang dari 10000 kasus diketahui tercatat sebanyak 661 hari
