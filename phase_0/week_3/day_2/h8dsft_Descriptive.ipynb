```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv("covid_19_indonesia_time_series_all.csv")
```

```python
df.head()
```

### 1. Cari Measure of Central Tendency dari data tersebut

```python
# cara 1
df.describe()
```

```python
# cara 2
mean = df["Total Cases"].mean()
median = df["Total Cases"].median()
mode = df["Total Cases"].mode()

print("Mean Total Cases adalah ", mean )
print("Median Total Cases adalah ", median)
print("Modus Total Cases adalah ", mode)
```

### 2. Cari Measure of Variance dari data tersebut.

```python
variance = df["Total Cases"].var()
std = df["Total Cases"].std()

print("Variance Total Cases adalah ", variance)
print("Standar deviasi Total Cases adalah ", std)
```
